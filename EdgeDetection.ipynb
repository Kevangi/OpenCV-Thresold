```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread(os.path.join('.', 'img', 'parrot.jpg'), cv2.IMREAD_GRAYSCALE)

# Step 4: Apply Gaussian Blur to reduce noise
blurred_image = cv2.GaussianBlur(image, (5, 5), 1.4)

# Step 5: Apply Canny edge detector
edges = cv2.Canny(blurred_image, 50, 150)

# Step 6: Display the original image and the edge-detected image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Edge Detected Image')
plt.imshow(edges, cmap='gray')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist

# Load MNIST dataset
(X_train, _), (_, _) = mnist.load_data()

# Create an output folder for edge-detected images
output_folder = "mnist_edges"
os.makedirs(output_folder, exist_ok=True)

# Process entire dataset and store first 10 edge images for visualization
edge_images = []
for i, img in enumerate(X_train):
    edges = cv2.Canny(img, 100, 200)  # Apply Canny edge detection
    
    # Save processed image
    output_path = os.path.join(output_folder, f"edge_{i}.png")
    cv2.imwrite(output_path, edges)
    
    # Store first 10 edge images for display
    if i < 10:
        edge_images.append(edges)

# Display the first 10 original and edge-detected images
fig, axes = plt.subplots(2, 10, figsize=(15, 5))

for i in range(10):
    # Original Image
    axes[0, i].imshow(X_train[i], cmap='gray')
    axes[0, i].set_title(f"Original {i+1}")
    axes[0, i].axis("off")

    # Edge-Detected Image
    axes[1, i].imshow(edge_images[i], cmap='gray')
    axes[1, i].set_title(f"Edges {i+1}")
    axes[1, i].axis("off")

plt.tight_layout()
plt.show()

print(f"Edge detection completed for MNIST dataset! Processed images saved in '{output_folder}'")

```
