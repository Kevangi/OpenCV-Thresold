```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist  # ✅ Corrected import

# Load Fashion-MNIST dataset
(X_train, _), (_, _) = fashion_mnist.load_data()  # ✅ Corrected dataset loading

# Select an image (first image in dataset)
img = X_train[0]

# Convert to 8-bit format for OpenCV
img = img.astype(np.uint8)

# 1️⃣ Gaussian Blurring (Smooth the image)
gaussian_blur = cv2.GaussianBlur(img, (5, 5), 0)

# 2️⃣ Sharpening (Enhance edges)
sharpening_kernel = np.array([[0, -1, 0],
                              [-1, 5, -1],
                              [0, -1, 0]])
sharpened = cv2.filter2D(img, -1, sharpening_kernel)

# 3️⃣ Sobel Edge Detection
sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined = cv2.magnitude(sobel_x, sobel_y)

# 4️⃣ Laplacian Edge Detection
laplacian = cv2.Laplacian(img, cv2.CV_64F)

# 5️⃣ Adaptive Thresholding
adaptive_thresh = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
                                        cv2.THRESH_BINARY, 11, 2)

# Display results
plt.figure(figsize=(12, 8))

# Original Image
plt.subplot(2, 3, 1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis("off")

# Gaussian Blurring
plt.subplot(2, 3, 2)
plt.imshow(gaussian_blur, cmap='gray')
plt.title("Gaussian Blur")
plt.axis("off")

# Sharpened Image
plt.subplot(2, 3, 3)
plt.imshow(sharpened, cmap='gray')
plt.title("Sharpened")
plt.axis("off")

# Sobel Edge Detection
plt.subplot(2, 3, 4)
plt.imshow(sobel_combined, cmap='gray')
plt.title("Sobel Edge Detection")
plt.axis("off")

# Laplacian Edge Detection
plt.subplot(2, 3, 5)
plt.imshow(laplacian, cmap='gray')
plt.title("Laplacian Edge Detection")
plt.axis("off")

# Adaptive Thresholding
plt.subplot(2, 3, 6)
plt.imshow(adaptive_thresh, cmap='gray')
plt.title("Adaptive Thresholding")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# 🔹 Set path to the folder containing images
dataset_path = "car/"  

# 🔹 Get list of images in the folder
image_files = [f for f in os.listdir(dataset_path) if f.endswith(('.png', '.jpg', '.jpeg'))]

# 🔹 Process first 5 images in the dataset
for idx, img_name in enumerate(image_files[:5]):
    img_path = os.path.join(dataset_path, img_name)

    # Load the image
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  # Convert to grayscale

    # 1️⃣ Gaussian Blurring
    gaussian_blur = cv2.GaussianBlur(img, (5, 5), 0)

    # 2️⃣ Sharpening
    sharpening_kernel = np.array([[0, -1, 0],
                                  [-1, 5, -1],
                                  [0, -1, 0]])
    sharpened = cv2.filter2D(img, -1, sharpening_kernel)

    # 3️⃣ Sobel Edge Detection
    sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
    sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
    sobel_combined = cv2.magnitude(sobel_x, sobel_y)

    # 4️⃣ Laplacian Edge Detection
    laplacian = cv2.Laplacian(img, cv2.CV_64F)

    # 5️⃣ Adaptive Thresholding
    adaptive_thresh = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
                                            cv2.THRESH_BINARY, 11, 2)

    # Display results
    plt.figure(figsize=(12, 8))

    # Original Image
    plt.subplot(2, 3, 1)
    plt.imshow(img, cmap='gray')
    plt.title(f"Original: {img_name}")
    plt.axis("off")

    # Gaussian Blurring
    plt.subplot(2, 3, 2)
    plt.imshow(gaussian_blur, cmap='gray')
    plt.title("Gaussian Blur")
    plt.axis("off")

    # Sharpened Image
    plt.subplot(2, 3, 3)
    plt.imshow(sharpened, cmap='gray')
    plt.title("Sharpened")
    plt.axis("off")

    # Sobel Edge Detection
    plt.subplot(2, 3, 4)
    plt.imshow(sobel_combined, cmap='gray')
    plt.title("Sobel Edge Detection")
    plt.axis("off")

    # Laplacian Edge Detection
    plt.subplot(2, 3, 5)
    plt.imshow(laplacian, cmap='gray')
    plt.title("Laplacian Edge Detection")
    plt.axis("off")

    # Adaptive Thresholding
    plt.subplot(2, 3, 6)
    plt.imshow(adaptive_thresh, cmap='gray')
    plt.title("Adaptive Thresholding")
    plt.axis("off")

    plt.tight_layout()
    plt.show()
```
